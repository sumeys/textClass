```
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import logging
from sklearn.metrics import f1_score, accuracy_score,classification_report

import wandb

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

train_df = pd.read_csv('/content/gdrive/MyDrive/fenlei/train_df1.csv')
train_df.columns = ["labels","text"]

eval_df = pd.read_csv('/content/gdrive/MyDrive/fenlei/eval_df1.csv')
eval_df.columns = ["labels","text"]

# dev_df = pd.read_csv('/content/gdrive/MyDrive/fenlei/dev_df.csv')
# dev_df.columns = ["labels","text"]
```

```
train_data = [
    ["科学规划", 1],
    ["提高技术附加值和品牌附加值", 2],
    ["提升品牌营销层次", 3],
    ["分清组织机构责任", 4],
    ["增强品牌保护", 5],
]
train_df = pd.DataFrame(train_data)
train_df.columns = ["text", "labels"]

# Preparing eval data
eval_data = [
    ["品牌规划", 1],
    ["采用新技术", 2],
    ["加大营销力度", 3],
    ["规范组织机构", 2],
    ["制定保护细则", 3],
]
eval_df = pd.DataFrame(eval_data)
eval_df.columns = ["text", "labels"]
```

```
train_df
```

```
model_args = ClassificationArgs()
model_args.num_train_epochs=20
model_args.learning_rate = 1e-4
model_args.labels_list = [1, 2,3, 4,5]

model = ClassificationModel(
    'bert',
    'bert-base-chinese',
    num_labels=5,
    args=model_args
) 
```

```
model.train_model(train_df)
```

```
from sklearn.metrics import precision_score,recall_score

def f1_multiclass(labels, preds):
      return precision_score(labels,preds,average='macro'),recall_score(labels,preds,average='macro'),f1_score(labels, preds, average='macro')
result, model_outputs, wrong_predictions = model.eval_model(eval_df, f1=f1_multiclass, acc=accuracy_score)
print(result)
```

```
predictions, raw_outputs = model.predict(["压实各方监管责任","提升质量安全保障","优化种植结构、推广农业新科技","推进农产品地理标志商标注册使用和保护工作","建立本地农产品在浙江直销网络"])
print(predictions, raw_outputs)
```
